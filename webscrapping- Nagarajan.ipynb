# importing libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
```

# website response

```python
res=requests.get(url)
```

```python
res
```

# parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
title_names=soup.select("td:nth-child(2)")
```

```python
title_names
```

```python
author_names=soup.select("td:nth-child(3)")
```

```python
author_names
```

```python
volume_sales=soup.select("td:nth-child(4)")
```

```python
volume_sales
```

```python
publisher=soup.select("td:nth-child(5)")
```

```python
publisher
```
